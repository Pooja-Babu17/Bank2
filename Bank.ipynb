```python
def freqTable(columnName,df):
    import pandas as pd
    frq=pd.DataFrame()
    frq["Unique_Values"]=dataset[columnName].value_counts().index
    frq["Frequency"]=dataset[columnName].value_counts().values
    frq["Relative_Fre"]=dataset[columnName].value_counts().values/len(dataset[columnName])*100
    frq["Cumulative"]=frq["Relative_Fre"].cumsum()
    return frq
```

```python
# data visual
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#time line
#How does my account balance change over time?
BankDataset['Date'] = pd.to_datetime(BankDataset['Date'])
plt.figure(figsize=(12, 6))
plt.plot(BankDataset['Date'], BankDataset['Balance'])
plt.title('Balance Over Time')
plt.xlabel('Date')
plt.ylabel('Balance')
plt.grid()
plt.show()
```

```python
BankDataset['Date'] = pd.to_datetime(BankDataset['Date'])
plt.figure(figsize=(12, 6))
plt.plot(BankDataset['Date'], BankDataset['Credit'])
plt.title('Credit Over Time')
plt.xlabel('Date')
plt.ylabel('Credit')
plt.grid()
plt.show()
```

```python
BankDataset['Date'] = pd.to_datetime(BankDataset['Date'])
plt.figure(figsize=(12, 6))
plt.plot(BankDataset['Date'], BankDataset['Debit'])
plt.title('Debit Over Time')
plt.xlabel('Date')
plt.ylabel('Debit')
plt.grid()
plt.show()
```

```python
#distribution of debit and credit amounts in my transactions
debit_df = BankDataset[BankDataset['DR/CR'] == 'DR']
credit_df = BankDataset[BankDataset['DR/CR'] == 'CR']

# Set up the subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))

# Plot histograms for debit amounts
axes[0].hist(debit_df['Debit'], bins=20, color='red', alpha=0.7, label='Debit')
axes[0].set_title('Debit Amount Distribution')
axes[0].set_xlabel('Amount')
axes[0].set_ylabel('Frequency')
axes[0].legend()

# Plot histograms for credit amounts
axes[1].hist(credit_df['Credit'], bins=20, color='blue', alpha=0.7, label='Credit')
axes[1].set_title('Credit Amount Distribution')
axes[1].set_xlabel('Amount')
axes[1].set_ylabel('Frequency')
axes[1].legend()

plt.show()

# Create box plots for debit and credit amounts
plt.figure(figsize=(10, 6))
sns.boxplot(x='DR/CR', y='Debit', data=BankDataset, palette=['red', 'blue'])
plt.title('Box Plot of Debit and Credit Amounts')
plt.xlabel('Transaction Type')
plt.ylabel('Amount')
plt.show()



```

```python
debit_data = BankDataset[BankDataset['DR/CR'] == 'DR']

# Group and sum debit transactions by the 'Remark' column
debit_remark_totals = debit_data.groupby('Remark')['Debit'].sum().reset_index()

# Create a scatter plot between 'Debit' and 'Remark' totals
plt.figure(figsize=(12, 6))
plt.scatter(debit_remark_totals['Debit'], debit_remark_totals['Remark'], color='purple', alpha=0.5)
plt.title('Scatter Plot of Debit vs. Remark Totals')
plt.xlabel('Total Debit Amount')
plt.ylabel('Remark')
plt.grid()
plt.show()
```

```python
credit_data =  BankDataset[ BankDataset['DR/CR'] == 'CR']
credit_remark_totals = credit_data.groupby('Remark')['Credit'].sum().reset_index()
plt.figure(figsize=(12, 6))
plt.scatter(credit_remark_totals['Credit'], credit_remark_totals['Remark'], color='blue', alpha=0.5)
plt.title('Scatter Plot of Credit vs. Remark Totals')
plt.xlabel('Total Credit Amount')
plt.ylabel('Remark')
plt.grid()
plt.show()
```
